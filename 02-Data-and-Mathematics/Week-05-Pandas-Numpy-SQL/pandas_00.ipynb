In pandas every thing round about two main data structure
1) Series

## The Series:
* A series is a one Dimension array(1D) or you can say it is an enhance python list.
### Difference:
* The main differnce between python list and pandas series.
#### 1)Pandas Series:
* pandas series has a label index it can be a number , date , string , productID 
#### 2)python list:
* It has a numeric Index



#### Example (Pandas Series)

```python
# It is a simple list of Temperature

temp_list = [22,23,45,67,90]

# Now we convert the temp_list it into pandas series

temperature = pd.Series(temp_list)
print(temperature)
```

* On the left side we notice that automatically index is generated 0,1,2,3 this is the functionality of pandas

### 2) DataFrame (2D)
It is a two Dimensional Data structure mean a table with row and column, It look like an excel sheet or SQl Table. A very common way to create a DataFrame from scratch is to use python Dictionary

### Example (DataFrame)

```python
# In Dictionary we have key and values(List) key--> Column List --> row

data = {'City':['London','Karachi','India','USA','Germany'],'Temperature':[24,34,40,14,32]}

# Now we Convert the dictionary into a Data Frame
df = pd.DataFrame(data)
print(df)
```

### Why We Use Pandas (Looping VS Broadcasting)
#### python list:
* We create a simple python list, Now if we want to add 5 in the entire list we need to use loop and go to the single element to add 5 
* Here is the Example 

```python
#Traditional way 

# Now we create a price list

price = [2,4,6,7,8,9]

for i in range(len(price)):
    print("The price after adding 2 at index",i ," is : ", price[i])
```

#### DataFrame (Broadcasting)
* We can create a simple python dictionary then converted it into python dataframe as df, Now if I want to add element in the entire list it not uses loop to do it, It simply applies the operation on the entire series at the hardware level
* * Example

```python
# Modern way
# Now first we create a price Dictionary

price = {'Product':['apple','Banana','Orange'],'Cost':[1,0.5,0.7]}

df = pd.DataFrame(price)
print(df)
updated_price = df*2

print("\n\n\n",updated_price)
```
